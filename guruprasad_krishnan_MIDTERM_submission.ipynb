```python
#quesion 1 

#initialize
cost =0.00;
result=[];
amount=0.00;
balance =0.00;

#item list
print("Item     Cost");
print("1.item_A  .75$");
print("2.item_B  1.45$");
print("3.item_C  4.30$");
print("4.item_D  .98$");
print("5.item_E  2.27$");
item_num = int(input("Input item number: ")); #get item to buy
if item_num==1:
    cost = 0.75;
elif item_num==2:
    cost = 1.45;
elif item_num==3:
    cost = 4.30;
elif item_num==4:
    cost = 0.98;
else:
    cost = 2.27;

# result aray
result = [];
amount = float(input("Insert amount(5 dollar bills only, here only 1 bill, assumption): "));
if amount != 5:
    print("Enter 5 dollar bill only"); #Insert More Money. question contradicts assumption of 5 dollar bills
else:
    balance = round(amount - cost,2);
    while balance>0.00:
        if balance >= 1.00:
            result.append(1);
            balance -= 1;
            continue;
        elif balance >= 0.25:
            result.append(0.25);
            balance -= 0.25;
            continue;
        elif balance >= 0.10:
            result.append(0.10);
            balance -= 0.10;
            continue;
        else:
            result.append(0.1);
            balance -= 0.1;
            continue;
print("-------------------");
print();
print("Item cost ",cost);
print("total balance ",round(amount-cost,2));
print("The change returned are as follows:");
print(result);
        
        
        
```

```python
#question 2

input_string = input("Enter String: ");
result = '';
for ch in input_string:
    if ord(ch) == 32:
        result += chr(45);
    else:
        result += chr(ord(ch)-32);
    
print("Output String:",result);
```
